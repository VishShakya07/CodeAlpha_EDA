```python
# Further feature extraction from df
import pandas as pd, re, matplotlib.pyplot as plt, seaborn as sns

df = df.copy()
# numeric price
df['Price'] = df['Prices']
# rating
df['Rating'] = pd.to_numeric(df['Reviews'], errors='coerce')
# brand from first word before space
df['Brand'] = df['Product Name'].str.split().str[0].str.upper()
# RAM GB
df['RAM_GB'] = df['Description'].str.extract('(\d+)\s*GB RAM', expand=False).astype(float)
# Battery mAh
df['Battery_mAh'] = df['Description'].str.extract('(\d{3,5})\s*mAh', expand=False).astype(float)
# Main camera MP (first occurrence of MP)
df['Camera_MP'] = df['Description'].str.extract('(\d{2,3})MP', expand=False).astype(float)
print(df[['Brand','RAM_GB','Battery_mAh','Camera_MP']].head())
print(df[['RAM_GB','Battery_mAh','Camera_MP']].describe())
# brand counts
brand_counts = df['Brand'].value_counts().head(10)
print(brand_counts)
# plot price vs RAM
plt.figure(figsize=(6,4))
sns.scatterplot(x='RAM_GB', y='Price', data=df)
plt.title('Price vs RAM')
plt.show()
# boxplot price by brand top5
top5 = brand_counts.index[:5]
plt.figure(figsize=(8,5))
sns.boxplot(x='Brand', y='Price', data=df[df['Brand'].isin(top5)])
plt.title('Price Distribution of Top 5 Brands')
plt.show()
```

```python
# Relationship between Rating and specs
import pandas as pd, seaborn as sns, matplotlib.pyplot as plt, statsmodels.api as sm

df = df.copy()
# ensure numeric cols exist
df['Price'] = df['Prices']
df['Rating'] = pd.to_numeric(df['Reviews'], errors='coerce')
df['RAM_GB'] = df['Description'].str.extract('(\d+)\s*GB RAM', expand=False).astype(float)
df['Battery_mAh'] = df['Description'].str.extract('(\d{3,5})\s*mAh', expand=False).astype(float)
df['Camera_MP'] = df['Description'].str.extract('(\d{2,3})MP', expand=False).astype(float)

numeric_cols = ['RAM_GB','Battery_mAh','Camera_MP','Price']
correlations = df[numeric_cols + ['Rating']].corr()[['Rating']].drop('Rating')
print(correlations)

# scatterplots
fig, axes = plt.subplots(1,3, figsize=(15,4))
for ax, col in zip(axes, ['RAM_GB','Battery_mAh','Camera_MP']):
    sns.regplot(x=col, y='Rating', data=df, ax=ax, scatter_kws={'alpha':0.6})
    ax.set_title(col + ' vs Rating')
plt.tight_layout()
plt.show()

# simple multivariate regression
X = df[numeric_cols]

# Drop rows with missing values in X and corresponding rows in df['Rating']
X = X.dropna()
y = df.loc[X.index, 'Rating']

X = sm.add_constant(X)

model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
# Price-performance analysis – build a simple performance index and visualize value
import pandas as pd, numpy as np, seaborn as sns, matplotlib.pyplot as plt
sns.set(style='ticks')

# use existing clean_df
df['Rating'] = pd.to_numeric(df['Reviews'], errors='coerce')
df['Price'] = df['Prices']
df['RAM_GB'] = df['Description'].str.extract('(\d+)\s*GB RAM', expand=False).astype(float)
df['Battery_mAh'] = df['Description'].str.extract('(\d{3,5})\s*mAh', expand=False).astype(float)
df['Camera_MP'] = df['Description'].str.extract('(\d{2,3})MP', expand=False).astype(float)

# keep rows with all numeric data
perf_df = df.dropna(subset=['Price','RAM_GB','Battery_mAh','Camera_MP','Rating']).copy()

# Normalize specs 0-1 (min-max)
for col in ['RAM_GB','Battery_mAh','Camera_MP','Rating']:
    perf_df[col + '_norm'] = (perf_df[col] - perf_df[col].min()) / (perf_df[col].max() - perf_df[col].min())

# Simple performance score: equally weighted average of four norms
perf_df['PerfScore'] = perf_df[['RAM_GB_norm','Battery_mAh_norm','Camera_MP_norm','Rating_norm']].mean(axis=1)

# Value metric: PerfScore per ₹10k
perf_df['Value'] = perf_df['PerfScore'] / (perf_df['Price']/10000)

# Identify top 10 value phones
top_value = perf_df.nlargest(10, 'Value')[['Product Name','Price','PerfScore','Value']]

print('Top 10 price-performance phones:')
print(top_value)

# Scatter plot Price vs. PerfScore
plt.figure(figsize=(7,5))
scatter = sns.scatterplot(data=perf_df, x='Price', y='PerfScore', hue='Value', palette='viridis', s=60)
plt.title('Price vs. Performance Score')
plt.xlabel('Price (₹)')
plt.ylabel('Performance Score (0-1)')
plt.legend(title='Value')
plt.tight_layout()
plt.show()
```

```python
# Histogram of Value scores and Price for all phones
import seaborn as sns, matplotlib.pyplot as plt
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
sns.histplot(perf_df['Price'], bins=30, kde=True, color='steelblue')
plt.xlabel('Price (₹)')
plt.title('Distribution of Phone Prices')
plt.subplot(1,2,2)
sns.histplot(perf_df['Value'], bins=30, kde=True, color='darkorange')
plt.xlabel('Value Score')
plt.title('Distribution of Value Scores')
plt.tight_layout()
plt.show()
```

```python
# Identify best budget phones using earlier perf_df
import pandas as pd, seaborn as sns, matplotlib.pyplot as plt

budget_cut = 10000
budget_df = perf_df[perf_df['Price'] <= budget_cut]

best_budget = budget_df.nlargest(10, 'Value')[['Product Name','Price','PerfScore','Value']]

print('Top budget phones (<= 10k):')
print(best_budget)

# Bar chart of value
plt.figure(figsize=(8,4))
plt.barh(best_budget['Product Name'], best_budget['Value'])
plt.xlabel('Value score')
plt.title('Best budget phones (<= ₹10k)')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```
